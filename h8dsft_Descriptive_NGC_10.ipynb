# Measure of Central Tendency 
 1. MEAN
 2. MEDIAN 
 3. MODUS
 

```python
obj_cols= list(df.select_dtypes('object'))
num_cols= list(df.select_dtypes('number'))
print(obj_cols)
print(num_cols)
```

```python
data_mean = df[num_cols].mean().rename('mean')
data_median = df[num_cols].median().rename('median')
data_mode = df[num_cols + obj_cols].mode().iloc[0].rename('mode')

central_ten = pd.concat([data_mean,data_median,data_mode], axis =1 )
central_ten
```

# Measure of Variance


1.   Standar deviasi
2.   Varians



```python
standar_dev = df[num_cols].std().rename('Standart_deviation')
varians = df[num_cols].var().rename('Varians')

Measure_of_variance = pd.concat([standar_dev, varians],axis=1)
Measure_of_variance
```

```python
complete_data = pd.concat([central_ten, Measure_of_variance], axis = 1)
complete_data
```
