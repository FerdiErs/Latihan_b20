Carilah gradient dari sebuah fungsi:<br>
3X + 4Y = 5 <br>

GRADIEN <br>
3/4 X + Y = 5<br>
Y = -3/4X + 5/4<br>

```python
import numpy as np
```

# Gradien X

```python
### define x
X = np.linspace(-50,50)

### define y
Y = 1.25 - 0.75 * X
```

```python
f = np.gradient(Y)
f
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,Y, color="Green")
plt.title('Plot X')

plt.show()
```

```python
plt.plot(f,Y, color = "pink")
plt.title('Gradient X')
plt.xlim(-2, 2)


plt.show()
```

# Gradien Y

Carilah gradient dari sebuah fungsi:<br>
3X + 4Y = 5

GRADIEN
X = 5/3 - 4/3X

```python
### define y
Y = np.linspace(-50,50)

### define x 
X =  (5/3)- (4/3) * X
```

```python
f1 = np.gradient(X)
f1
```

```python
plt.plot(X,Y, color="red")
plt.title('Plot Y')

plt.show()
```

```python
plt.plot(X,f1, color = "pink")
plt.title('Gradient Y')

plt.show()
```
